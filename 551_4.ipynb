```
import requests
from bs4 import BeautifulSoup
```

```
#Get the data (already in a friendly format)
CR_raw = requests.get("https://raw.githubusercontent.com/jacobsanz97/551-proj4/master/CR.txt")
CR = BeautifulSoup(CR_raw.content, "html.parser")

MPQA_raw = requests.get("https://raw.githubusercontent.com/jacobsanz97/551-proj4/master/MPQA.txt")
MPQA = BeautifulSoup(MPQA_raw.content, "html.parser")

SST2_raw = requests.get("https://raw.githubusercontent.com/jacobsanz97/551-proj4/master/SST2.txt")
SST2 = BeautifulSoup(SST2_raw.content, "html.parser")
```

```
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer

#Can specify "vocabulary" attribute to restrict the vocabulary to word2vec
tfidf3_vec = TfidfVectorizer(ngram_range=(1,3), norm='l2', smooth_idf=True)
count3_vec = CountVectorizer(ngram_range=(1,3))


def split(X, y, vec):
  X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
  vec.fit(X_train)
  X_train = vec.transform(X_train)
  X_test = vec.transform(X_test)
  return X_train, X_test, y_train, y_test

def count(X, y):
  return (split(X, y, count3_vec))

def tfidf(X, y):
  return (split(X, y, tfidf3_vec))
  
```

```
# models to test
from sklearn.ensemble import VotingClassifier, RandomForestClassifier, AdaBoostClassifier
from sklearn.naive_bayes import MultinomialNB
from sklearn.tree import DecisionTreeClassifier
from sklearn import linear_model, svm

#Logistic regression
regr_model = linear_model.LogisticRegression(solver = 'lbfgs', penalty='l2', random_state=seed,
                                            tol=1e-4, C=1.0)#hyperparam to grid search


#Linear SVM
svm_model = svm.LinearSVC(random_state=seed)

#Polynomial SVM
poly_model = svm.SVC(kernel='poly', cache_size=500, degree=2, random_state=seed)

#The multinomial distribution normally requires integer feature counts. However, in practice, fractional counts such as tf-idf may also work.
#from documentation
#Multinomial Naive Bayes
mnb_model = MultinomialNB(alpha=1)

#Random Forest Classifier
rfc_model = RandomForestClassifier(n_estimators=50, random_state=seed)

#AdaBoost
abc_model = AdaBoostClassifier(DecisionTreeClassifier(max_depth=1, random_state=seed),
                         algorithm="SAMME", n_estimators=500)

#Stacking
estimators = [('Logistic Regression', regr_model), ('Linear SVM', svm_model), ('Multinomial NB', mnb_model),
             #('Random Forest', rfc_model), ('AdaBoost', abc_model)
             ]
vot_model = VotingClassifier(estimators, voting='hard')
```

```
from sklearn.metrics import f1_score, accuracy_score

#Train and predict for models (unsure if only works on sklearn model)
#Prints F1 and accuracy on training and validation sets
def test_model(model, X_train, X_test, y_train, y_test):
  model.fit(X_train, y_train)
  y_train_p = model.predict(X_train)
  y_test_p = model.predict(X_test)
  score_model(y_train, y_test, y_train_p, y_test_p)
  
#Prints F1 measure and accuracy for the model, on training and test set
#Args: y_train, y_test (ground truth), y_train_p, y_test_p (predictions)
def score_model(y_train, y_test, y_train_p, y_test_p):
  f1_train = f1_score(y_train, y_train_p)
  f1_test = f1_score(y_test, y_test_p)
  acc_train = accuracy_score(y_train, y_train_p)
  acc_test = accuracy_score(y_test, y_test_p)
  print("Training set: F1 measure: ", f1_train, " accuracy: ", acc_train)
  print("Validation set: F1 measure: ", f1_test, " accuracy: ", acc_test)
```
